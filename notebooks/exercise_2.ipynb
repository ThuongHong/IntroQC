```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

## Bài 2

### a. $\ket{\psi_1} = \frac{\sqrt{3}}{2}\ket{0} + \frac{1}{2}\ket{1}$

$$
    \ket{\psi_1} = \frac{\sqrt{3}}{2}\ket{0} + \frac{1}{2}\ket{1}= \cos\frac{\pi}{6}\ket{0} + e^{i0}\sin\frac{\pi}{6}\ket{1}\\
    \implies \theta = \frac{\pi}{3}, \phi = 0
$$

```python
theta = np.pi/3
phi = 0

# [r, theta, phi]
plot_bloch_vector([1, theta, phi], title="θ = π/3, φ = 0", coord_type='spherical')
```

# b. $\ket{\psi_2} = \frac{1}{\sqrt{2}}(\ket{0} + e^{i\frac{\pi}{6}}\ket{1})$
$$
    \ket{\psi_2} = \frac{1}{\sqrt{2}}\ket{0} + \frac{1}{\sqrt{2}}e^{i\frac{\pi}{6}}\ket{1} = \cos\frac{\pi}{4}\ket{0} + e^{i\frac{\pi}{6}}\sin\frac{\pi}{4}\ket{1}\\
                 \implies \theta = \frac{\pi}{2}, \phi = \frac{\pi}{6}
$$

```python
theta = np.pi/2
phi = np.pi/6

# [r, theta, phi]
plot_bloch_vector([1, theta, phi], title="θ = π/2, φ = π/6", coord_type='spherical')
```

```python
theta = 2 * np.arccos(2/3)
phi = np.arctan(-2)

# [r, theta, phi]
plot_bloch_vector([1, theta, phi], title="θ = 2 * arccos(2/3), φ = arctan(-2)", coord_type='spherical')
```

## Bài 4

### Vẽ mạch mô tả tính toán HYTHX với trạng thái đầu vào $\ket{0}$

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.y(0)
qc.t(0)
qc.h(0)
qc.x(0)

qc.draw(output='mpl')
```

```python
# Trạng thái input (mặc định là |0>)
psi_in = Statevector.from_int(0, 2**1) # 2**1 = 1 qubit

psi_out = psi_in.evolve(qc)

print(psi_out)
```

```python
qc = QuantumCircuit(1, 1)

qc.h(0)
qc.y(0)
qc.t(0)
qc.h(0)
qc.x(0)

qc.measure(0, 0)

qc.draw(output='mpl', initial_state=True)
```
